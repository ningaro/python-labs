### Задача 1

Даны две строки: `s1` и `s2` с одинаковым размером. 

Проверьте, может ли некоторая перестановка строки `s1` “победить” некоторую перестановку строки `s2` или наоборот.

Строка `x` может “победить” строку `y` (обе имеют размер `n`), если `x[i] >= y[i]` (в алфавитном порядке) для всех `i` от 0 до `n-1`.

```python
# Функция возвращает массив со всеми возможными комбинациями
def permutation(pa):
    res = []
    _permutation(pa, 0, res)
    return res

# Функция добавляет в массив с результатоми новые комбинации 
def _permutation(pa, i, res):
    if i == len(pa) - 1:
        elem = ''
        for a in pa:
            elem += str(a)
        if elem not in res:
            res.append(elem)
    else:
        for j in range(i, len(pa)):
            swap(pa, i, j)
            _permutation(pa, i + 1, res)
            swap(pa, i, j)

# Функция изменящая местами элементы в массиве (из конца в начало)
def swap(pa, i, j):
    k = pa[i]
    pa[i] = pa[j]
    pa[j] = k

# Основное тело программы
s1 = input() # Вводим строку S1
s2 = input() # Вводим строку S2
if len(s1) == len(s2):
    list_s1 = list(s1); list_s2 = list(s2)
    perms_s1 = permutation(list_s1)
    perms_s2 = permutation(list_s2)
    res_s1 = []; res_s2 = []
    for i in perms_s1:
        for j in perms_s2:
            c = True
            for k in range(len(s1)):
                if ord(i[k]) < ord(j[k]):
                    c = False
            if c and i not in res_s1:
                res_s1.append(i)
    for i in perms_s2:
        for j in perms_s1:
            c = True
            for k in range(len(s1)):
                if ord(i[k]) < ord(j[k]):
                    c = False
            if c and i not in res_s2:
                res_s2.append(i)
    print(len(res_s1) > 0 or len(res_s2) > 0)
else:
    print("Длинны строк должны совпадать!")
```

### Задача 2

Дана строка `s`, вернуть самую длинную полиндромную подстроку в `s`.

```python
# Функция возвращает перевернутую строку
def reversed(str):
    return str[::-1]

# Основное тело программы
s = input()
res = ''
for i in range(len(s)-2):
    for j in range(i+2, len(s)+1):
        if s[i:j] == reversed(s[i:j]):
            if len(s[i:j]) > len(res):
                res = s[i:j]
            elif len(s[i:j]) == len(res):
                res = s[i:j]
if res == '':
    print("Тут неть палиндромов!") 
else: 
    print(res)
```

### Задача 3

Вернуть количество отдельных непустых подстрок текста, которые могут быть записаны как конкатенация некоторой строки с самой собой (т.е. она может быть записана, как `a + a`, где `a` - некоторая строка).

```python
s = input()
res = ['']
for i in range(len(s)-2):
    for j in range(i+2, len(s)+1):
        if (j - i) % 2 == 0:
            center = (i + j) // 2
            if s[i:center] == s[center:j] and s[i:j] not in res:
                if len(s[i:j]) > len(res[0]):
                    res = [s[i:j]]
                elif len(s[i:j]) == len(res[0]):
                    res.append(s[i:j])
if res[0] == '': 
    print("Таких подстрок нету!")
else:
    print(len(res))
```
